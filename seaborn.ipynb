```python
# 2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

np.random.seed(42)
data = np.random.randn(1000)

sns.set(style="whitegrid")
sns.histplot(data, kde=True)

plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.


data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [23, 45, 56, 78, 43]
}

df = pd.DataFrame(data)

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.barplot(x="Category", y="Value", data=df, palette="Blues_d")

plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.


np.random.seed(42)
data = {
    'Category': np.random.choice(['A', 'B', 'C', 'D'], size=200),
    'Value': np.random.randn(200) * 20 + 100 
}
df = pd.DataFrame(data)

plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Boxplot of Value by Category')


plt.subplot(1, 2, 2)
sns.violinplot(x='Category', y='Value', data=df)
plt.title('Violinplot of Value by Category')

plt.tight_layout()
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(42)
size = 300

X1 = np.random.normal(0, 1, size)
X2 = 0.8 * X1 + np.random.normal(0, 1, size)
X3 = 0.5 * X1 - 0.3 * X2 + np.random.normal(0, 1, size)
X4 = np.random.normal(0, 1, size)
X5 = np.random.normal(0, 1, size)

df = pd.DataFrame({
    'Feature_1': X1,
    'Feature_2': X2,
    'Feature_3': X3,
    'Feature_4': X4,
    'Feature_5': X5
})

corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
